### Между какими событиями наибольший шаг? Укажите пару идентификаторов событий с наибольшим шагом, относительно всей базы (всех файлов).


```python
query = ''' WITH tab as
            (SELECT event_id,
                   session_id,
                   request_time,
                   LEAD(event_id) OVER(PARTITION BY session_id ORDER BY request_time) as next_event,
                   LEAD(request_time) OVER(PARTITION BY session_id ORDER BY request_time) as next_time
              FROM events_all)
              SELECT *,
              ROUND((JULIANDAY(next_time) - JULIANDAY(request_time)) * 86400, 5) as event_duration
              FROM tab
              ORDER BY event_duration DESC            
        '''
event_data1 = pd.io.sql.read_sql(query, con = con12)
event_data1
```

Наибольший шаг за три дня зафиксирован между событиями 6 "Начинаем читать данные из BQ по полученному SQL запросу" и 7 "Получили данные из BQ по полученному SQL запросу" в сессии 4dcab2af4573ced8251fb7433174ba15. Он составил более 106 секунд.            

### Интервал времени до наступления какого события показывает наибольший разброс. Укажите пару идентификатор и sender с наибольшим разбросом. Предложите свои идей, с чем это может быть связано.


```python
query = ''' WITH tab as
            (SELECT event_id,
                   session_id,
                   sender,
                   request_time,
                   LEAD(event_id) OVER(PARTITION BY session_id ORDER BY request_time) as next_event,
                   LEAD(request_time) OVER(PARTITION BY session_id ORDER BY request_time) as next_time
              FROM events_all),
              tab1 as
              (SELECT *,
              ROUND((JULIANDAY(next_time) - JULIANDAY(request_time)) * 86400, 5) as event_duration
              FROM tab
              ORDER BY event_duration DESC)
             
             SELECT 
                    sender,
                    next_event,
                    MAX(event_duration) - MIN(event_duration) as spread_of_values
             FROM tab1
             WHERE event_duration > 0
             GROUP BY next_event, sender
             ORDER BY spread_of_values DESC
        '''
event_data2 = pd.io.sql.read_sql(query, con = con12)
event_data2
```

Наибольший разброс в значениях имеет интервал времени до наступления события 7 "Получили данные из BQ по полученному SQL запросу" при постановке задачи weekly_report.care. Он составляет более 105 секунд. Вероятно на это влияет размер считываемых данных: чем больший объем считываем, тем дольше выполняется операция.

### Какое количество задач прогнозирования могут выполняться одновременно? Укажите максимальное число параллельных задач.

```python
query = ''' SELECT session_id,
                   min(request_time) as session_start,
                   max(request_time) as session_end
              FROM events_all
          GROUP BY session_id          
        '''
event_data4 = pd.io.sql.read_sql(query, con = con12)
event_data4
```

```python
# создадим массив времен и отсортируем его
times = pd.concat([event_data4['session_start'], event_data4['session_end']])\
        .sort_values().reset_index(drop=True).rename('time')

# переведем в формат хранения времени
times = pd.to_datetime(times)

# создаем вспомогательный столбец следующих времен, обновляем индекс и объединяем в датафрейм
times1 = pd.Series(times[1:], name='next_time').reset_index(drop=True)
times_df = pd.concat([times, times1], axis=1)

# расчитываем среднее между двумя точками времени
times_df['delta'] = times_df['time'] + (times_df['next_time'] - times_df['time'])/2
times_df
```

```python
times_df.info()
```

```python
event_data4['session_end'] = pd.to_datetime(event_data4['session_end'])
event_data4['session_start'] = pd.to_datetime(event_data4['session_start'])
event_data4.info()
```

```python
session_count = []
for i in tqdm(range(times_df.shape[0])):
    sc = 0
    for j in range(event_data4.shape[0]):
        if (event_data4['session_start'][j] <= times_df['delta'][i]) & (event_data4['session_end'][j] >= times_df['delta'][i]):
            sc +=1
    session_count.append(sc)
    #session_count[i] = sc
max(session_count)
```

Алгоритм отрабатывает очень долго. Попробуем применить библиотеку pandas.

```python
session_count = []
for i in tqdm(range(times_df.shape[0])):
    ev_df1=event_data4[event_data4['session_start'] <= times_df['delta'][i]]
    ev_df2=ev_df1[ev_df1['session_end'] >= times_df['delta'][i]]
    sc = ev_df2.shape[0]
    session_count.append(sc)
    #session_count[i] = sc
max(session_count)
```

Максимальное количество задач, обрабатываемых одновременно - 6.

Альтернативное решение (проверим себя). Создадим датафрейм со всеми временными метками и флагом начало\конец выполнения задачи.

```python
d1={'time': event_data4['session_start']}
df1 = pd.DataFrame(data=d1)
df1['start'] = 1
df1.head()
```

```python
d2={'time': event_data4['session_end']}
df2 = pd.DataFrame(data=d2)
df2['start'] = 0
df2.head()
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df = df.sort_values(by='time').reset_index(drop=True)
df.head()
```

Если задача начинается прибавим 1, если задача закончена вычтем 1. 

```python
%%time
task_count = 0
max_count = 0
for i in range(df.shape[0]):
    if df['start'][i] == 1:
        task_count +=1
    else:
        task_count -=1
    if task_count > max_count:
        max_count = task_count
max_count
```
